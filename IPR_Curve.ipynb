```python
#Importing the essential libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#The Pwf and q of Reservoir Pressure at 2000, 1500, and 1000 are entered
Pwf_2000 = [2000, 1800, 1600, 1400, 1200, 1000, 800, 600, 400, 200, 0]
q_2000 = [0, 380, 720, 1020, 1280, 1500, 1680, 1820, 1920, 1980, 2000]

Pwf_1500 = [1500, 1350, 1200, 1050, 900, 750, 600, 450, 300, 150, 0]
q_1500 = [0, 160, 304, 430, 540, 633, 709, 768, 813, 835, 844]

Pwf_1000 = [1000, 900, 800, 700, 600, 500, 400, 300, 200, 100, 0]
q_1000 = [0, 48, 90, 128, 160, 188, 210, 228, 240, 248, 250]

#The title, y-axis and x-axis are entered
plt.title("IPR Curves for Different Reservoir Pressures")
plt.ylabel("Pwf(psig)")
plt.xlabel('q(Stb/Day)')

#The actual plots are done here
plt.plot(q_2000, Pwf_2000,'o-', label = "Reservoir Pressure = 2000psig")
plt.plot(q_1500, Pwf_1500, 'o-', label = "Reservoir Pressure = 1500psig")
plt.plot(q_1000, Pwf_1000, 'o-', label = "Reservoir Pressure = 1000psig")
plt.legend()
```
